```python
# Lets calculate the functions of the these images.

hist = cv2.calcHist([leo], channels=[0], mask=None, histSize=[256], ranges=[0,256])
hist.shape   # (256,1)
plt.plot(hist)
```

```python
# Plot all color Histograms.

img = leo
color = ('b', 'g', 'r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])

plt.title('Histogram of Leopard')
```

# Part 2 : Histogram Equalization.


```python
# Notes: 

# Histgram Eq is a method of contrast adjustment based on the images hist.
# Matrix ---> Zoom In Pic ---> Histogram Analysis.
# When we apply Hist Eq we get high contrast on our image.

rain = cv2.imread("rain.png")
show = cv2.cvtColor(rain, cv2.COLOR_BGR2RGB)

img = rain
mask = np.zeros(img.shape[:2])  # 1 color cause its all zeros.
plt.imshow(mask)
```

```python
# Slicing our mask.

mask[300:400,100:400] = 255
plt.imshow(mask)
```

```python
mask_rain = cv2.bitwise_and(img, img, mask=mask)
show = cv2.bitwise_and(rain, rain, mask=mask)
plt.imshow(img)
```

```python
downs = cv2.imread('downs.png')

def display(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
    
display(downs)
```

```python
downs.shape
```

```python
hist_downs = cv2.calcHist([downs], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_downs)
```

```python
# Lets EQ the hist. Increasing the contrast.

eq = cv2.equalizeHist(downs)
display(eq, cmap='gray')
```

```python
hist_downs = cv2.calcHist([eq], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_downs)
```
