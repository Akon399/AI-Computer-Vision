```python
# Blending & Pasting images.

img1 = cv2.imread('SF.png')
img2 = cv2.imread('NY.png')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# Fix the two images to be off original color.

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# Resize the images to be the same size.

img1 = cv2.resize(img1,(700,700))
img2 = cv2.resize(img2,(700,700))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# Blend the images togehter.

blend = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.imshow(blend)   # We get equal rep of both images.
```

```python
# Pasting a small img on a large img.

img1 = cv2.resize(img1,(2000,2000))
large = img1
small = img2
```

```python
x0 = 0
y0 = 0
```

```python
x1 = x0 + small.shape[1]
y1 = y0 + small.shape[0]
```

```python
large[y0:y1,x0:x1] = small
```

```python
plt.imshow(large)
```
