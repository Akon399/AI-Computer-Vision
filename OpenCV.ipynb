```python
# Image = 0's = 1 colour.
blank = np.zeros(shape=(512,512,3), dtype=np.int16)
blank
```

```python
# Show the color of Image = 0's.
plt.imshow(blank)
```

```python
# Drawing a rectangle on an image.
# What does drawing help with? It helps with Object Detection, our Machine must be able to identify objects & name them.

cv2.rectangle(blank, pt1=(300,0), pt2=(400,400), color=(0,250,0), thickness=10)
plt.imshow(blank)
```

```python
# Draw a Square.

cv2.rectangle(blank, pt1=(100,100), pt2=(300,300), color=(10,10,250), thickness=10)
plt.imshow(blank)
```

```python
# Drawing Circles.

cv2.circle(blank, center=(400,400), radius=100, color =(100,0,0), thickness=10)
plt.imshow(blank) # These circles are for cneters = (100,100) & (500,500) & (400,400).
```

```python
# For drawing a line we use , Line = cv2.line().
# Writing text on images.

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank, text="Hi Akhona :-), hows Computer Vision Research?", org=(10,500),
           fontFace=font, fontScale=2, color=(200,200,200), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank)
```
